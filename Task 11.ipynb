```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("apps.csv")
df
```

```python
print("Initial Data Overview:")
print(df.info())
print(df.head())
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True)
df['Price'] = df['Price'].str.replace('[$]', '', regex=True)
```

```python
df
```

```python
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df.dropna(inplace=True)

```

```python
category_counts = df['Category'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=category_counts.index[:10], y=category_counts.values[:10], palette='viridis')
plt.xticks(rotation=45)
plt.xlabel("App Category")
plt.ylabel("Number of Apps")
plt.title("Top 10 App Categories in Google Play Store")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], bins=20, kde=True, color='blue')
plt.xlabel("App Rating")
plt.ylabel("Count")
plt.title("Distribution of App Ratings")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Price'])
plt.xlabel("Price ($)")
plt.title("Price Distribution of Apps")
plt.show()
```

```python
df_filtered = df[(df['Installs'] > 0) & (df['Rating'] > 0)]
plt.figure(figsize=(10,6))
sns.scatterplot(x=df_filtered['Installs'], y=df_filtered['Rating'], alpha=0.6)
plt.xscale('log')
plt.xlabel("Number of Installs (Log Scale)")
plt.ylabel("App Rating")
plt.title("Relationship between Installs and Rating")
plt.show()

print("Data Analysis and Visualization Completed.")
```
